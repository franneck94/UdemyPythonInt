```python
import math
import time
from threading import Thread
```

# GIL: Global Interpreter Lock

```python
numbers = [
    1102023313711321,
    2102023313556361,
    2133132415198713,
    2456241020233131,
    2102124102331313,
    2102023151513137,
    2102023313112511,
]
```

```python
def is_prime(n: int) -> bool:
    if n < 2:
        return False
    if n in (2, 3, 5, 7, 11, 13, 17):
        return True
    if (
        n % 2 == 0
        or n % 3 == 0
        or n % 5 == 0
        or n % 7 == 0
        or n % 11 == 0
        or n % 13 == 0
        or n % 17 == 0
    ):
        return False
    upper_limit = int(math.sqrt(n)) + 1
    for i in range(19, upper_limit, 2):
        if n % i == 0:
            return False
    return True
```

```python
start = time.perf_counter_ns()

for number in numbers:
    is_prime(number)

end = time.perf_counter_ns()
print(f"time: {(end - start) / 1000000.0} ms")
```

```python
threads = [Thread(target=is_prime, args=(number,)) for number in numbers]
```

```python
start = time.perf_counter_ns()

[t.start() for t in threads]
[t.join() for t in threads]

end = time.perf_counter_ns()
print(f"time: {(end - start) / 1000000.0} ms")
```
